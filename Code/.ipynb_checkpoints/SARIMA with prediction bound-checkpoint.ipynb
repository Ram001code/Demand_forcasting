### All in one 

```python
data_df= data2.copy()
```

```python
data_df
```

```python
# train
```

```python
for i in data_df['Location'].unique():
    df= data_df[data_df['Location']==i]
    for j in df['Item'].unique():
        item_data= df[df["Item"]==j]
        
        item_data= item_data.drop(['Location', 'Item', 'Amount'], axis=1)
        time_series = item_data.index
        train = item_data.iloc[:len(df)-31]
        test = item_data.iloc[len(df)-31:] # set 31 Months for testing
        
        from statsmodels.tsa.statespace.sarimax import SARIMAX
  
        model = SARIMAX(train['Quantity'], 
                order = (1, 1, 1), 
                seasonal_order =(2, 1, 1, 12))
  
        result = model.fit()
    
        forecast = result.predict(start = 1, 
                          end = (len(item_data)), 
                          typ = 'levels').rename('Forecast')
        
        result_df = (result.get_prediction(start='2020-01-05', end= len(item_data)+31).summary_frame(alpha=0.75))
        
        df1= pd.DataFrame()
        df1['Date'] = result_df.index
        df1['Forecast_Qty']= result_df['mean'].values
        df1['Pred_lower']= result_df['mean_ci_lower'].values
        df1['Pred_upper']= result_df['mean_ci_upper'].values
        
        #Creating df for storing data without lag. 
        stock= pd.DataFrame()
        stock['Date'] = time_series
        stock['Forecasted_Qty']= forecast.values


#         df_forecast = pd.DataFrame()
#         df_forecast['Date']= time_series
#         df_forecast['Forcasted_Qty']= forecast.values
        
        #Saving Results 
        df1['Location']= i
        df1['Item']=j
        
        stock['Location']= i
        stock['Item']=j
        
        print("\n")
        print(f'Writing prediction bound results of {i} and {j}')
        predction_output = "ModelOut\All_out\Weekly\Output_bound\Output_bound -" +i+"-"+j+".xlsx"
        df1.to_excel(predction_output, index= False)
        
        print(f'Writing stk data for {i} and {j}')
        predction_output = "ModelOut\All_out\Weekly\Output\Output_stk-" +i+"-"+j+".xlsx"
        stock.to_excel(predction_output, index= False)
        
        print('\n')
        print('Calculating error matrices')
        
        from sklearn.metrics import mean_squared_error
#         from statsmodels.tools.eval_measures import rmse

        from sklearn.metrics import mean_absolute_percentage_error

        import math

        MSE_Score= mean_squared_error(item_data["Quantity"], forecast.values).round(2)
#         print('MSE_Score:', MSE_Score)

        RMSE_Score = math.sqrt(mean_squared_error(item_data["Quantity"], forecast.values))
#         print('RMSE_Score:', round(RMSE_Score, 2))

        MAPE_Score= mean_absolute_percentage_error(item_data["Quantity"], forecast.values).round(2)
#         print('MAPE_Score:', MAPE_Score)
        
        col_names =  ['MSE_SARIMA', 'RMSE_SARIMA', 'MAPE_SARIMA']
        Error_matric= pd.DataFrame(columns = col_names)
        
        my_out = {'MSE_SARIMA': MSE_Score, 'RMSE_SARIMA': RMSE_Score, 'MAPE_SARIMA': MAPE_Score }
        Error_matric.loc[len(Error_matric)] = my_out 
        
        Error_matric['Location']=i
        Error_matric['Item']=j
        
        print(f'Writing error matrices of {i} and {j}')
        print("\n")
        Error_output = "ModelOut\All_out\Weekly\Error\Output_Error -" +i+"-"+j+".xlsx"
        Error_matric.to_excel(Error_output, index= False)
        
        print('Training complete')
        
```
