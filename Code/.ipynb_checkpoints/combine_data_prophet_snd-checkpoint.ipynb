```python
df = data[["Date","Location","Item","Quantity"]]
df
```

```python
lakehood = df[df["Location"]=="LAKEWOOD"]
lakehood
```

```python
lakehood.Item.unique()
```

```python
lakehood_chkn = lakehood[lakehood["Item"]=="CHICKEN XXXXX - O"]
lakehood_chkn
```

```python
gb_chicken = lakehood_chkn.groupby("Date")["Quantity"].sum().reset_index().sort_values(by="Date")
gb_chicken.head()
```

```python
gb_chicken.rename(columns={"Date": "ds", "Quantity": "y"}, inplace=True)
gb_chicken.tail()
```

```python
# Modelling - Prophet
m = Prophet().fit(gb_chicken)
future = m.make_future_dataframe(periods=30, freq='D')
pred = m.predict(future)
```

```python
try_gb = gb_chicken.copy()
try_gb["yhat_lower"] = try_gb["y"]
try_gb["yhat_upper"] = try_gb["y"]
try_gb.rename(columns={"y":"yhat"}, inplace=True)

a = [try_gb,pred[["ds", "yhat", "yhat_lower", "yhat_upper"]]]
pred_new = pd.concat(a).reset_index()

past = pred_new["ds"].iloc[-100:-30]
future = pred_new["ds"].tail(31)

past_s = pred_new["yhat"].iloc[-100:-30]
future_s = pred_new["yhat"].tail(31)

fig, ax = plt.subplots(figsize=(15,8))

ax.set_title("LAKEHOOD - CHICKEN XXXXX - O")
ax.set_ylabel('Quantity')
ax.set_xlabel('Date')
ax.plot(past, past_s, color='dodgerblue')
ax.plot(future, future_s, color='crimson')
ax.fill_between(future, pred_new["yhat_lower"].tail(31), pred_new["yhat_upper"].tail(31), color='crimson', alpha=0.3)
z = i+"-"+j+" Plot"
plt.savefig(str(z), bbox_inches='tight', dpi=150)
```

### Plots and prediction df for each location

```python
for i in data["Location"].unique():
    df = data[data["Location"]==i]
    for j in df["Item"].unique():
        itm_data = df[df["Item"]==j]
        gb = itm_data.groupby("Date")["Quantity"].sum().reset_index().sort_values(by="Date")
        gb.rename(columns={"Date": "ds", "Quantity": "y"}, inplace=True)
        
        #modelling
        m = Prophet().fit(gb)
        future = m.make_future_dataframe(periods=30, freq='D')
        pred = m.predict(future)
        
        try_gb = gb.copy()
        try_gb["yhat_lower"] = try_gb["y"]
        try_gb["yhat_upper"] = try_gb["y"]
        try_gb.rename(columns={"y":"yhat"}, inplace=True)

        a = [try_gb,pred[["ds", "yhat", "yhat_lower", "yhat_upper"]]]
        pred_new = pd.concat(a).reset_index()

        past = pred_new["ds"].iloc[-100:-30]
        future = pred_new["ds"].tail(31)

        past_s = pred_new["yhat"].iloc[-100:-30]
        future_s = pred_new["yhat"].tail(31)

        fig, ax = plt.subplots(figsize=(15,8))
        
        a = i+"-->"+j
        
        ax.set_title(a)
        ax.set_ylabel('Quantity')
        ax.set_xlabel('Date')
        ax.plot(past, past_s, color='dodgerblue')
        ax.plot(future, future_s, color='crimson')
        ax.fill_between(future, pred_new["yhat_lower"].tail(31), pred_new["yhat_upper"].tail(31), color='crimson', alpha=0.3)
        
        z = "Plot "+i+"-"+j
        plt.savefig(str(z), bbox_inches='tight', dpi=150)
        
        df_cv = cross_validation(m, initial='60 days', period='100 days', horizon = '30 days')
        
        df_p = performance_metrics(df_cv)
        
        df_p["Location"] = i
        df_p["Item"] = j
        
        pred["Location"] = i
        pred["Item"] = j
        
        predction_output = "Output -" +i+"-"+j+".xlsx"
            
        Error_metrics = "Error_metrics -"+i+"-"+j+".xlsx"
        
        pred.to_excel(predction_output)
        
        df_p.to_excel(Error_metrics)
        
        
        
```
