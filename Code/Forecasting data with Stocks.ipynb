```python
conditions = [
    (data_forct['stock_percentage'] <= 0),
    (data_forct['stock_percentage'] > 0) & (data_forct['stock_percentage'] <= 25),
    (data_forct['stock_percentage'] > 25)]
```

```python
values = ['Understock', 'Onhand', 'Overstock']
data_forct["stock"]=np.select(conditions, values)
```

```python
len(data_forct[data_forct['stock']== 'Onhand'])
```

```python
def stock_flag(row):
    if row["stock"]== 'Overstock':
        return 1
    elif row["stock"]== 'Understock':
        return 2
    if row["stock"]== 'Onhand':
        return 3
```

```python
data_forct["stock_flag"] = data_forct.apply(stock_flag, axis=1)
```
