```python
data_merge = pd.merge(df_weekly1,df_sales, on=['Date', 'Location', 'Item'], how='inner')
```

```python
data_merge
```

```python
data_merge['Quantity'] = data_merge['Quantity'].round(2)
data_merge['Amount'] = data_merge['Amount'].round(2)
data_merge['Forecasted_Qty'] = data_merge['Forecasted_Qty'].round(2)
```

```python
data_merge.head()
```

```python
# Stocks = MAX(0, D(LT) - On hand - InTrn(LT) + PndSales(LT)).

data_merge['Stock_Qty'] = round(data_merge['Forecasted_Qty'] +data_merge['On Order Quantity'] - data_merge['OH Quantity']-data_merge['In Transit Quantity'], 0) 
```

```python
data_merge
```

```python
data_merge1= data_merge[data_merge['Date'] >='2022-04-27']
```

```python
# data_merge1.to_excel('6 months inventory.xlsx', index= False)
```

```python
data_merge1['Stock_Qty'].mean()
```

```python
data_merge1['Stock_Qty'].std()
```

```python
conditions = [
    (data_merge1['Stock_Qty'] <= 0),
    (data_merge1['Stock_Qty'] > 0) & (data_merge1['Stock_Qty'] <= 30),
    (data_merge1['Stock_Qty'] > 30)]

values = ['Understock', 'Onhand', 'Overstock']

data_merge1["Stocks"]=np.select(conditions, values)
```

```python
data_merge1[data_merge1['Stocks']== 'Overstock']
```

```python
def stock_flag(row):
    if row['Stocks']== 'Overstock':
        return 1
    elif row['Stocks']== 'Understock':
        return 2
    elif row['Stocks']=='Onhand':
        return 3
    
    
```

```python
data_merge1["Stock_flag"] = data_merge1.apply(stock_flag,axis=1)
```

### Analysing Sales data 

```python
df_sales
```

```python
df_sales1 = df_sales[df_sales['Date'] >='2022-04-27']
df_sales1
```

```python
df_sales1['Quantity'] = df_sales1['Quantity'].round(2)
df_sales1['Amount'] = df_sales1['Amount'].round(2)
df_sales1['Forecasted_Qty'] = df_sales1['Forecasted_Qty'].round(2)
```

```python
df_sales1
```

```python
df_sales1["stock_percentage"] = ((df_sales1["Forecasted_Qty"]-df_sales1["Quantity"])/df_sales1["Quantity"])*100
```

```python
df_sales1
```

```python
conditions = [
    (df_sales1['stock_percentage'] <= 0),
    (df_sales1['stock_percentage'] > 0) & (df_sales1['stock_percentage'] <= 20),
    (df_sales1['stock_percentage'] > 20)]
```

```python
values = ['Understock', 'Onhand', 'Overstock']
df_sales1["Stock"]=np.select(conditions, values)
```

```python
def stock_flag(row):
    if row["Stock"]== 'Overstock':
        return 1
    elif row["Stock"]== 'Understock':
        return 2
    if row["Stock"]== 'Onhand':
        return 3
```

```python
df_sales1["Stock_flag"] = df_sales1.apply(stock_flag,axis=1)
```
