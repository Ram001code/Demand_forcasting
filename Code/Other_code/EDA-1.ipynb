```python
#Filtering product items only 

product_item = ['FOOD', 'FRESH FOOD', 'SOFTLINES', 'HARDLINES']

filter_product = filter1[filter1['MAJOR_DEPT_GRP_LONG_DESC'].isin(product_item)]
```

```python
data_rename = filter_product.rename(columns={'CALENDAR_YR_NUM': 'Year', 'CALENDAR_MO_NUM': 'Month', 'COMPANY_CODE': 'Cntry', \
                             'BI_RPT_STATE_CODE': 'State_Code', "BI_RPT_REGION_DESC": "Region", 'BI_RPT_DISTRICT_DESC': "Dist_Code", \
                             "LOCATION_LONG_NAME": "Location", "MAJOR_DEPT_GRP_LONG_DESC": "Major Department"})
```

```python
data_rename = data_rename.rename(columns={'BI_DEPARTMENT_NUM_LDESC': 'Department', "BI_MAJOR_CATEGORY_CODE_DESC": "Major_Category", \
                                            "BI_GROUP_CATEGORY_CODE_DESC":  "Grp_ctgry", "BI_ITEM_CATEGORY_CODE_DESC": "Item_ctgry", \
                                            "ITEM_DESC_1": "Item"})
```

```python
data_rename
```

```python
data_rename.columns
```

```python
data_reindex = data_rename.reindex(columns= ['Year', 'Month', 'year_month', 'Data_type', 'Cntry', 'Region', 'State_Code', \
                               'Dist_Code', 'Location', 'Major Department', 'Department', 'Major_Category', \
                               'Grp_ctgry', 'Item_ctgry', 'Item', 'OnHand_QTY', 'OnHand_AMT', 'OnOrder_QTY', \
                               'OnOrder_AMT', 'InTransit_QTY', 'InTransit_AMT', 'Non-Saleable-Items_QTY', \
                               'Non-Saleable-Items_AMT'])
```
