```python
qt= rfm_tb.quantile(q=[0.25,0.5,0.75])
qt
```

```python
qt= qt.to_dict()
qt
```

```python
rfm_seg = rfm_tb.copy()
```

```python

def R_seg(a,b,c):
    if a<=c[b][0.25]:
        return 1
    elif a<=c[b][0.50]:
        return 2
    elif a<=c[b][0.75]:
        return 3 
    else:
        return 4

def FM_seg(a,b,c):
    if a<=c[b][0.25]:
        return 4
    elif a<=c[b][0.50]:
        return 3
    elif a<=c[b][0.75]:
        return 2 
    else:
        return 1
    
```

```python
rfm_seg['R_Quartile'] = rfm_seg['Recency'].apply(R_seg, args=('Recency', qt,))
rfm_seg['F_Quartile'] = rfm_seg['Frequency'].apply(FM_seg, args=('Frequency', qt,))
rfm_seg['M_Quartile'] = rfm_seg['Monetary'].apply(FM_seg, args=('Monetary', qt,))
```

```python
rfm_seg['RFM_Score'] = rfm_seg['R_Quartile'].map(str)+ rfm_seg['F_Quartile'].map(str)\
                        + rfm_seg['M_Quartile'].map(str)
```

```python
rfm_seg.head()
```

```python
rfm_seg['Total_Score']= rfm_seg['R_Quartile'] + rfm_seg['F_Quartile'] + rfm_seg['M_Quartile']
rfm_seg.head()
```
