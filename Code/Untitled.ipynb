```python
## Aploading Sales Inventory Weekly data

df_sales = pd.read_excel('C:\\Users\\Ramshankar\\OneDrive - iLink Systems Inc\\Documents\\Costco\\Dataset\\Final Output\\Weekly\\SARIMA\\Final\\All_Output_SARIMA_without_bound_weekly.xlsx')
```

```python
data_merge = pd.merge(data_inv,df_sales, on=['Date', 'Location', 'Item'], how='inner')
```

```python
data_merge
```

```python
data_merge['Quantity'] = data_merge['Quantity'].round(2)
data_merge['Amount'] = data_merge['Amount'].round(2)
data_merge['Forecasted_Qty'] = data_merge['Forecasted_Qty'].round(2)
```

```python
data_merge.tail()
```

```python
# 6 months filtering 

data_merge1= data_merge[data_merge['Date'] >='2022-04-27']
```

```python
data_merge1.head()
```

```python
data_merge1.tail()
```

```python
# Stocks = MAX(0, D(LT) - On hand - InTrn(LT) + PndSales(LT)).

data_merge1['Stock_Qty'] = round(data_merge1['Quantity'] +data_merge1['On Order Quantity'] - data_merge1['OH Quantity']-data_merge1['In Transit Quantity'], 0) 
```

```python
conditions = [
    (data_merge1['Stock_Qty'] <= 0),
    (data_merge1['Stock_Qty'] > 0) & (data_merge1['Stock_Qty'] <= 20),
    (data_merge1['Stock_Qty'] > 20)]

values = ['Understock', 'Onhand', 'Overstock']

data_merge1["Stocks"]=np.select(conditions, values)
```

```python
len(data_merge1[data_merge1['Stocks']== 'Onhand'])
```

```python
#Overstock-222
#Understock-144
#Onhand-42
```

```python
def stock_flag(row):
    if row['Stocks']== 'Overstock':
        return 1
    elif row['Stocks']== 'Understock':
        return 2
    elif row['Stocks']=='Onhand':
        return 3
    
```

```python
data_merge1["Stock_flag"] = data_merge1.apply(stock_flag,axis=1)
```
