```python
## Aploading Sales Inventory Weekly data

df_frct= pd.read_excel('C:\\Users\\Ramshankar\\OneDrive - iLink Systems Inc\\Documents\\Costco\\Dataset\\Final Output\\Weekly\\SARIMA\\Final\\All_Output_SARIMA_without_bound_weekly.xlsx')
```

```python
df_weekly1['Item'] = df_weekly1['Item'].replace('Chicken XXXXX-O', 'CHICKEN XXXXX - O')
```

```python
df_weekly1['Item']== 'CHICKEN XXXXX - O'
```

```python
data_merge= pd.merge(df_weekly1,df_frct, on=['Date', 'Location', 'Item'], how='inner')
```

```python
data_merge
```

```python
# Stocks = MAX(0, D(LT) - On hand - InTrn(LT) + PndSales(LT)).

data_merge['Stock_Qty'] = round(data_merge['Forecasted_Qty'] +data_merge['On Order Quantity'] - data_merge['OH Quantity']
                                -data_merge['In Transit Quantity'], 0) 
```

```python
data_merge
```

```python
conditions = [
    (data_merge['Stock_Qty'] <= 0),
    (data_merge['Stock_Qty'] > 0) & (data_merge['Stock_Qty'] <= 25),
    (data_merge['Stock_Qty'] > 25)]

values = ['Understock', 'Onhand', 'Overstock']

data_merge["Stocks"]=np.select(conditions, values)
```

```python
data_merge[data_merge['Stocks']=='Understock']
```

```python
def stock_flag(row):
    if row['Stocks']== 'Overstock':
        return 1
    elif row['Stocks']== 'Understock':
        return 2
    elif row['Stocks']=='Onhand':
        return 3
    
```

```python
data_merge["Stock_flag"] = data_merge.apply(stock_flag,axis=1)
```
